```python
# Importing python libraries 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

### Importing Python Libraries

In this section, we import the essential Python libraries required for data analysis and visualization:

- **pandas** (`pd`): Used for data manipulation and analysis, especially with structured data (e.g., dataframes).
- **matplotlib.pyplot** (`plt`): A plotting library used for creating static, animated, and interactive visualizations in Python.
- **numpy** (`np`): Provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on them.


```python
#A dictionary with student data.
data = {
    'S_No': [101, 102, 103, 104, 105],
    'Stu_Name': ['Avni', 'Bhavya', 'Lubhanshi', 'Anshika', 'Soumya'],
    'Humanities': [75, 82, 68, 10, 78],
    'English': [88, 76, 12, 80, 85],
    'Physics': [65, 20, 85, 75, 90],
    'Chemistry': [38, 85, 70, 68, 82],
    'Maths': [92, 88, 78, 95, 18]
}
```

### Creating a Dictionary with Student Data

We define a Python dictionary named `data` containing information about five students. Each key in the dictionary represents a column or attribute of the dataset:

- `S_No`: Serial number of each student.
- `Stu_Name`: Names of the students.
- `Humanities`, `English`, `Physics`, `Chemistry`, `Maths`: Marks scored by each student in different subjects.

This structured dictionary format allows us to easily convert it into a DataFrame for further analysis and visualization.


```python
# Converting the data dictionary into a pandas DataFrame and display it
df = pd.DataFrame(data)
df
```

### Converting the Dictionary into a DataFrame

We convert the `data` dictionary into a **pandas DataFrame** using `pd.DataFrame()`. This creates a tabular structure where:

- Each key in the dictionary becomes a column header.
- Each list under a key becomes a column of data.

Displaying the DataFrame (`df`) allows us to view the student records in a structured, spreadsheet-like format, which is easier to analyze and visualize.

```python
# Adding a new column 'Total_Marks' to calculate the total subject marks
df['Total_Marks'] = df[['Humanities', 'English', 'Maths', 'Physics', 'Chemistry']].sum(axis=1)
df
```

### Calculating Total Marks for Each Student

We add a new column called `Total_Marks` to the DataFrame. This column stores the **sum of marks obtained by each student** across all subjects:

- `Humanities`
- `English`
- `Maths`
- `Physics`
- `Chemistry`

We use the `sum(axis=1)` function to calculate the row-wise total (i.e., for each student). Displaying the updated DataFrame now shows the total marks alongside individual subject scores.


```python
# Adding a new column 'Percentage' based on total marks to calculate percentage of each student.
df['Percentage'] = (df['Total_Marks'] / 500) * 100
df
```

### Calculating Percentage for Each Student

We add another column called `Percentage` to the DataFrame. This column represents the **percentage of total marks** each student scored out of 500 (assuming each subject is out of 100 and there are 5 subjects).

The percentage is calculated using the formula:

\[
\text{Percentage} = \left(\frac{\text{Total Marks}}{500}\right) \times 100
\]

This makes it easier to compare student performance in a normalized way.


```python
# Defines the grading and adds a Grade column based on percentage.
def get_grade(percentage):
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B'
    elif percentage >= 60:
        return 'C'
    elif percentage >= 40:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Percentage'].apply(get_grade)
df
```

### Assigning Grades Based on Percentage

We define a function `get_grade()` to assign letter grades based on the student's percentage:

- **A+**: 90 and above
- **A** : 80–89
- **B** : 70–79
- **C** : 60–69
- **D** : 40–59
- **F** : Below 40

We then use the `apply()` function to apply this grading logic to each student's `Percentage` value, creating a new column called `Grade` in the DataFrame.

This helps in categorizing student performance in a clear, standardized format.


```python
# Display grades of each student
for index, row in df.iterrows():
    print(f"{row['Stu_Name']} scored {row['Percentage']:.2f}% and received grade '{row['Grade']}'")
```

### Displaying Each Student's Grade

We iterate through each row of the DataFrame using `iterrows()` and print a formatted message that shows:

- The student's name (`Stu_Name`)
- Their calculated percentage (`Percentage`)
- The assigned grade (`Grade`)

This provides a readable summary of individual student performance in the console.


```python
#1.graph for total mark of each student
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
pastel_colors = ['#AEC6CF','#FFB347','#77DD77','#FF6961','#CBAACB']  

plt.bar(df['Stu_Name'], df['Total_Marks'], color=pastel_colors)
plt.title('Student-wise Total Marks')
plt.xlabel('Student Name')
plt.ylabel('Total Marks')
plt.xticks(rotation=45)
```

### Bar Graph: Total Marks of Each Student

We create a **bar chart** to visualize the total marks scored by each student. Key elements of the graph include:

- **X-axis**: Student names
- **Y-axis**: Total marks out of 500
- **Bar colors**: Soft pastel colors for visual appeal
- The chart is placed in the first position of a 2x2 subplot layout (preparing space for more plots later)
- Student names on the x-axis are rotated for better readability

This graph helps in quickly comparing the overall academic performance of the students.


```python
#2. line chart of performnce of each student in each subject 
subjects = ['Humanities', 'English', 'Physics', 'Chemistry', 'Maths']

markers = ['o', 's', '^', 'D', 'P'] 
plt.figure(figsize=(10, 6))

for i, row in df.iterrows():
    student_name = row['Stu_Name']
    marks = [row[subject] for subject in subjects]
    plt.plot(subjects, marks, marker=markers[i % len(markers)], label=student_name)

plt.title('Subject-wise Marks per Student')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.legend(title="Students")
plt.grid(True)
plt.suptitle("Each Student's Performance in Subjects", fontsize=14, y=1.05)
plt.tight_layout()
plt.show()
```

### Line Chart: Performance of Each Student Across Subjects

This line chart visualizes the marks scored by each student in different subjects:

- The **X-axis** represents the subjects (`Humanities`, `English`, `Physics`, `Chemistry`, `Maths`).
- The **Y-axis** shows the marks obtained in each subject.
- Each student’s performance is plotted as a separate line with unique markers for easy identification.
- A legend indicates which line corresponds to which student.
- Grid lines improve readability of the chart.

This chart helps to compare how each student performs across various subjects and identify their strengths and weaknesses.


```python
# 3.bar graph for classwise performance of each student in each subject 
bar_width = 0.15
index = np.arange(len(df['Stu_Name']))
colors = {'Humanities': '#FFB3BA',   'English': '#BAE1FF','Physics': '#BFFCC6','Chemistry': '#E2C2FF',    'Maths': '#FFDAC1'}

plt.bar(index, df['Humanities'], bar_width, label='Humanities', color=colors['Humanities'])     
plt.bar(index + bar_width, df['English'], bar_width, label='English', color=colors['English']) 
plt.bar(index + 2 * bar_width, df['Physics'], bar_width, label='Physics', color=colors['Physics']) 
plt.bar(index + 3 * bar_width, df['Chemistry'], bar_width, label='Chemistry', color=colors['Chemistry'])  
plt.bar(index + 4 * bar_width, df['Maths'], bar_width, label='Maths', color=colors['Maths'])  

plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.title('Class-wise Performance (Marks in Each Subject)')
plt.xticks(index + 2 * bar_width, df['Stu_Name'], rotation=45)
plt.legend()
plt.show()
```

### Grouped Bar Graph: Subject-wise Performance of Each Student

This grouped bar chart displays marks scored by each student in all subjects side-by-side for easy comparison:

- The **X-axis** lists the student names.
- The **Y-axis** shows the marks in each subject.
- Each subject is represented by a different colored bar.
- Bars for each student are grouped together to show all subject marks at once.
- The legend maps colors to subjects for clarity.
- Student names on the x-axis are rotated for better readability.

This visualization helps to analyze and compare student performance across all subjects in a single view.


```python
# 3.pie chart for classwise performance of each student in each subject 
def assign_grade(marks):
    if marks >= 90:
        return 'A+'
    elif marks >= 80:
        return 'A'
    elif marks >= 70:
        return 'B'
    elif marks >= 60:
        return 'C'
    elif marks >= 40:
        return 'D'
    else:
        return 'F'

subjects = ['Humanities', 'English', 'Physics', 'Chemistry', 'Maths']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']

for subject in subjects:
    plt.figure(figsize=(7, 7))
    
    df[subject] = pd.to_numeric(df[subject], errors='coerce')
    
    grades = df[subject].apply(assign_grade)
    
    grade_counts = grades.value_counts().reindex(['F', 'D', 'C', 'B', 'A', 'A+'], fill_value=0)
    
    wedges, texts, autotexts = plt.pie(
        grade_counts,
        labels=grade_counts.index,
        autopct='%1.1f%%',
        colors=colors,
        startangle=140
    )
    
    plt.title(f'Grade Distribution in {subject}')
    plt.legend(wedges, grade_counts.index, title="Grades", loc="center left", bbox_to_anchor=(1, 0.5))
    plt.show()
```

### Pie Charts: Grade Distribution in Each Subject

For each subject, we visualize the distribution of grades among students using pie charts:

- We first define the `assign_grade()` function to classify marks into letter grades (`A+`, `A`, `B`, `C`, `D`, `F`).
- For each subject, the marks are converted to numeric values (handling any errors).
- We apply the grading function to each student's marks to get their grade.
- We then count how many students fall into each grade category.
- A pie chart is created to show the percentage of students in each grade category per subject.
- Different colors distinguish the grades visually.
- A legend and title are included for clarity.

This set of charts helps to quickly understand the overall grade distribution and performance patterns in each subject.
